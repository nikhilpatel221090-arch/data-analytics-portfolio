```python
# Customer Complaints Dashboard - Organized Layout

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("cleaned_customer_complaints.csv")

# ===================== KPIs =====================
total_complaints = len(df)
unique_customers = df["CustomerID"].nunique()
avg_satisfaction = round(df["Satisfaction"].mean(), 2)
most_common_complaint = df["Complaint"].mode()[0]

# Show KPIs
print("ðŸ“Š KEY METRICS")
print(f"âœ… Total Complaints: {total_complaints}")
print(f"âœ… Unique Customers: {unique_customers}")
print(f"âœ… Avg Satisfaction Score: {avg_satisfaction}/5")
print(f"âœ… Most Common Complaint: {most_common_complaint}")

# ===================== Dashboard Layout =====================
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle("Customer Complaints Dashboard", fontsize=16, fontweight="bold")

# Chart 1: Complaints Frequency
df["Complaint"].value_counts().plot(kind="bar", ax=axs[0,0], color="orange")
axs[0,0].set_title("Complaints by Type")
axs[0,0].set_ylabel("Count")

# Chart 2: Complaint Distribution Pie
df["Complaint"].value_counts().plot(kind="pie", autopct="%1.1f%%", ax=axs[0,1])
axs[0,1].set_ylabel("")
axs[0,1].set_title("Complaint Distribution")

# Chart 3: Satisfaction Histogram
df["Satisfaction"].plot(kind="hist", bins=5, rwidth=0.8, color="purple", ax=axs[1,0])
axs[1,0].set_title("Satisfaction Score Distribution")
axs[1,0].set_xlabel("Score (1â€“5)")

# Chart 4: Complaints per Customer (Top 10)
df["CustomerID"].value_counts().head(10).plot(kind="bar", ax=axs[1,1], color="green")
axs[1,1].set_title("Top 10 Customers by Complaints")
axs[1,1].set_ylabel("Count")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
