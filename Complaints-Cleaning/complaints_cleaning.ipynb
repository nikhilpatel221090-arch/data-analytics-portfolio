```python
# Customer Complaints Data Cleaning

import pandas as pd

# Load the raw dataset
raw = pd.read_csv("raw_customer_complaints.csv")
print("ðŸ”¹ Raw Data Sample")
display(raw.head())

# Step 1: Check column names & data info
print("\nðŸ”¹ Dataset Info")
print(raw.info())

print("\nðŸ”¹ Missing Values Count")
print(raw.isnull().sum())

# Step 2: Remove duplicates
no_duplicates = raw.drop_duplicates()

# Step 3: Handle missing values (drop rows with nulls)
cleaned = no_duplicates.dropna()

# Step 4: Save cleaned dataset
cleaned.to_csv("cleaned_customer_complaints.csv", index=False)

print("\nâœ… Cleaning complete! Saved as cleaned_customer_complaints.csv")

# Show cleaned data preview
display(cleaned.head())
```
