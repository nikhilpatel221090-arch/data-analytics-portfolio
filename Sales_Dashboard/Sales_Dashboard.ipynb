```python
# Sales Performance Dashboard - Organized Layout

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data (1).csv")

# ===================== KPIs =====================
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
avg_profit_margin = round(total_profit / total_sales * 100, 2)

# Show KPIs
print("ðŸ“Š KEY METRICS")
print(f"âœ… Total Sales: {total_sales}")
print(f"âœ… Total Profit: {total_profit}")
print(f"âœ… Avg Profit Margin: {avg_profit_margin}%")

# ===================== Dashboard Layout =====================
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle("Sales Performance Dashboard", fontsize=16, fontweight="bold")

# Chart 1: Sales by Region
region_sales = df.groupby("Region")["Sales"].sum()
region_sales.plot(kind="bar", ax=axs[0,0], color="skyblue")
axs[0,0].set_title("Total Sales by Region")
axs[0,0].set_ylabel("Sales")

# Chart 2: Monthly Sales Trend
monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales.plot(kind="line", marker="o", ax=axs[0,1])
axs[0,1].set_title("Monthly Sales Trend")
axs[0,1].set_ylabel("Sales")

# Chart 3: Profit by Category
category_profit = df.groupby("Category")["Profit"].sum()
category_profit.plot(kind="bar", ax=axs[1,0], color="green")
axs[1,0].set_title("Profit by Category")
axs[1,0].set_ylabel("Profit")

# Chart 4: Profit vs Sales
axs[1,1].scatter(df["Sales"], df["Profit"], alpha=0.6, color="red")
axs[1,1].set_title("Profit vs Sales")
axs[1,1].set_xlabel("Sales")
axs[1,1].set_ylabel("Profit")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
