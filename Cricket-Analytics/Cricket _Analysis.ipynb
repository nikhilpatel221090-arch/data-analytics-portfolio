```python
# Cricket Analytics Dashboard - Organized Layout

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("cricket_data.csv")

# ===================== KPIs =====================
total_runs = df["Runs"].sum()
top_scorer = df.loc[df["Runs"].idxmax(), "Player"]
avg_strike_rate = round(df["StrikeRate"].mean(), 2)
best_avg_player = df.loc[df["Average"].idxmax(), "Player"]

# Show KPIs
print("ðŸ“Š KEY METRICS")
print(f"âœ… Total Runs: {total_runs}")
print(f"âœ… Top Scorer: {top_scorer}")
print(f"âœ… Avg Strike Rate: {avg_strike_rate}")
print(f"âœ… Best Batting Average: {best_avg_player}")

# ===================== Dashboard Layout =====================
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle("Cricket Analytics Dashboard", fontsize=16, fontweight="bold")

# Chart 1: Runs by Player
df.plot(x="Player", y="Runs", kind="bar", ax=axs[0,0], color="green", legend=False)
axs[0,0].set_title("Total Runs by Player")
axs[0,0].set_ylabel("Runs")

# Chart 2: Strike Rate by Player
df.plot(x="Player", y="StrikeRate", kind="bar", ax=axs[0,1], color="blue", legend=False)
axs[0,1].set_title("Strike Rate by Player")
axs[0,1].set_ylabel("Strike Rate")

# Chart 3: Batting Average
df.plot(x="Player", y="Average", kind="bar", ax=axs[1,0], color="orange", legend=False)
axs[1,0].set_title("Batting Average by Player")
axs[1,0].set_ylabel("Average")

# Chart 4: Run Contribution Pie
df.set_index("Player")["Runs"].plot(kind="pie", autopct="%1.1f%%", ax=axs[1,1])
axs[1,1].set_ylabel("")
axs[1,1].set_title("Run Contribution")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
